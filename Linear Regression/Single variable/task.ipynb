```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('canada_per_capita_income.csv')
data.head()
```

```python
%matplotlib inline
plt.scatter(data.per_capita_income,data.year)
```

```python
plt.scatter(data.per_capita_income,data.year,color='red',marker='+')
plt.xlabel('Income')
plt.ylabel('Year')
```

```python
data[['year']]
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(data[['year']],data.per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
plt.xlabel('Year')
plt.ylabel('Per_capita_income')
plt.scatter(data.year,data.per_capita_income,color='red',marker='+')
plt.plot(data.year,reg.predict(data[['year']]),color='blue')
```
